```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator as IDG
```

```python
tf.__version__
```

```python
gen_data= IDG(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)

training_data= gen_data.flow_from_directory("C:/Users/Lenovo/Desktop/dogs_cats/training_set",target_size=(64,64),batch_size=35,class_mode='binary')
```

```python
test_gen_data= IDG(rescale=1./255)
test_data= test_gen_data.flow_from_directory("C:/Users/Lenovo/Desktop/dogs_cats/test_set",target_size=(64,64),batch_size=35,class_mode='binary')
```

```python
#Building the CNN
#Initialization CNN
```

```python
cnn= tf.keras.models.Sequential()
```

```python
#ADD convolutional layer or input layer
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,activation='relu',input_shape=[64,64,3],kernel_size=3))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
#add second convolutional layer
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32,activation='relu',kernel_size=3))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
#add Flattening layer
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
#Full Connection
```

```python
cnn.add(tf.keras.layers.Dense(units=128,activation='relu'))
```

```python
#Output Layer
```

```python
cnn.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
#training the cnn
#Compiling the CNN
```

```python
cnn.compile(optimizer='adam',loss='binary_crossentropy', metrics=['accuracy'])
```

```python
#training the CNN
```

```python
cnn.fit(x=training_data,validation_data=test_data,epochs=20)
```

```python
import numpy as np
from keras.preprocessing import image
test_image= image.load_img("C:\\Users\\Lenovo\\Desktop\\dogs_cats\\single_prediction\\cat_dog9.jpg", target_size=(64,64))
test_image= image.img_to_array(test_image)
test_image= np.expand_dims(test_image,axis=0)
result= cnn.predict(test_image)
if result[0][0]==1:
    prediction='Dog'
else:
    prediction='Cat'
```
